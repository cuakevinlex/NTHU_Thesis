```python
#converts all midi files in the current folder
import json
import glob
import os
import music21

count = 0
#converting everything into the key of C major or A minor

#tranpose songs to -3 to 3
semitones = [-3, -2, -1, 0, 1, 2, 3]
failed =[]
#os.chdir("./")
cwd = os.getcwd()
folder = "POP909_augmented"
def transpose(semitones, folder, filename):
    for k in semitones:
        midi = MidiFile(cwd+'\\POP909\\'+filename+'\\'+filename+'.mid')
        createmidi = MidiFile()
        for track in midi.tracks:
            createtrack = MidiTrack()
            createtrack = track
            for msg in createtrack:
                if msg.type == "note_on":
                    msg.note = msg.note + k
            createmidi.tracks.append(createtrack)
        createfolder = cwd+'\\' + folder + '\\'+filename+'\\'
        Path(createfolder).mkdir(parents=True, exist_ok=True)
        createmidi.save(cwd+'\\' + folder + '\\'+filename+'\\'+filename + "_" + str(k) + ".mid")
with open('solos.json') as json_file: 
    data = json.load(json_file) 
for i in range(1,910):
    filename = str(i).zfill(3)
    if filename not in data:
        continue
    for j in range(len(data[filename])):
        count += 1
        try:
            for k in semitones:
            # extract song, then, tranpose, then save
                intro = music21.converter.parse("./data/melody/intro/" + str(i).zfill(3) + "_solo_" + str(j) + ".mid")
                i_score= intro.transpose(k)
                newfilename = "./data/data_augmented_melody/intro/" + str(i).zfill(3) + "_solo_" + str(j) + "_" + str(k) + ".mid"
                i_score.write("midi",newfilename)

                outro = music21.converter.parse("./data/melody/outro/" + str(i).zfill(3) + "_solo_" + str(j) + ".mid")
                o_score = outro.transpose(k)
                newfilename = "./data/data_augmented_melody/outro/" + str(i).zfill(3) + "_solo_" + str(j) + "_" + str(k) + ".mid"
                o_score.write("midi",newfilename)

                solo = music21.converter.parse("./data/melody/middle/" + str(i).zfill(3) + "_solo_" + str(j) + ".mid")
                s_score = solo.transpose(k)
                newfilename = "./data/data_augmented_melody/middle/" + str(i).zfill(3) + "_solo_" + str(j) + "_" + str(k) + ".mid"
                s_score.write("midi",newfilename)
        except:
            print("Error " + str(i).zfill(3))
            failed.append(str(i).zfill(3))
        print(count)
```
