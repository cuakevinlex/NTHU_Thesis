```python
from __future__ import print_function
import time
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

```python
cwd = os.getcwd()
files = os.listdir(cwd)  # Get all the files in that directory
# print("Files in %r: %s" % (cwd, files))
data = pd.read_excel (cwd+'/data.xlsx', engine='openpyxl')
# data = pd.read_excel (cwd+'/solo_iso.xlsx', engine='openpyxl')
print(data)
```

```python
df_intro = pd.DataFrame(data, columns= ['intro_noteseq'])
df_solo = pd.DataFrame(data, columns= ['middle_noteseq'])
df_outro = pd.DataFrame(data, columns= ['outro_noteseq'])

solo = df_solo.values
intro = df_intro.values
outro = df_outro.values
```

```python
intro_train, intro_subset, solo_train, solo_subset, outro_train, outro_subset = train_test_split(intro, solo, outro, test_size=226)

intro_val, intro_test, solo_val, solo_test, outro_val, outro_test = train_test_split(intro_subset, solo_subset, outro_subset, test_size=113)
```

```python
train = []
val = []
test = []

for i in range(len(intro_train)):
    train.append([intro_train[i][0],outro_train[i][0], solo_train[i][0]])
    
for i in range(len(intro_val)):
    val.append([intro_val[i][0],outro_val[i][0], solo_val[i][0]])
    
for i in range(len(intro_test)):
    test.append([intro_test[i][0],outro_test[i][0], solo_test[i][0]])
```

```python
df_train = pd.DataFrame(train,columns=['intro', 'outro', 'solo'])
df_val = pd.DataFrame(test,columns=['intro', 'outro', 'solo'])
df_test = pd.DataFrame(val,columns=['intro', 'outro', 'solo'])
```

```python
destination_folder="solo_generator_dataset_text"
df_train.to_csv(destination_folder + '/train.csv', index=False)
df_val.to_csv(destination_folder + '/val.csv', index=False)
df_test.to_csv(destination_folder + '/test.csv', index=False)
```
