```python
def find_min(input_path):
    lowest_note = 100
    for midi_file in input_path:
        mid = MidiFile(midi_file, clip=True)
        for track in mid.tracks:
            for msg in track:
                if msg.type =="note_on" and msg.note < lowest_note:
                    lowest_note = msg.note
    return lowest_note
```

```python
def print_notes(input_path):
    mid = MidiFile(input_path, clip=True)
    notes = []
    for track in mid.tracks:
        for msg in track:
            if msg.type=="note_on":
                print(msg)
    return notes
```

```python
def extract_events(input_path):
    mid = MidiFile(input_path, clip=True)
    notes = []
    for track in mid.tracks:
        for msg in track:
            if msg.type=="note_on" and msg.velocity != 0:
                notes.append(msg.note)  
    return notes
```

```python
def extract_set_events(input_path):
    mid = MidiFile(input_path, clip=True)
    notes_on = []
    notes_off = []
    notes = []
    for track in mid.tracks:
        for msg in track:
            if msg.type=="note_on":
                if msg.velocity == 0:
                    notes_off.append(msg.note)
                else:
                    notes_on.append(msg.note)
                notes.append(msg.note)
    set_on = set(notes_on)
    set_off = set(notes_off)
    notes = intersection(set_on, set_off)
    return list(set(notes))
```

```python
def extract_events_norm(input_path):
    mid = MidiFile(input_path, clip=True)
    notes = []
    for track in mid.tracks:
        for msg in track:
            if msg.type=="note_on" and msg.velocity != 0:
                notes.append(msg.note - min_note + 1)  
    return notes
```

```python
def extract_events_localmin_norm(input_path):
    mid = MidiFile(input_path, clip=True)
    notes = []
    local_min = 128
    unique_notes = extract_set_events(input_path)
    for note in unique_notes:
        if local_min > note:
            local_min = note
    for track in mid.tracks:
        for msg in track:
            if msg.type=="note_on" and msg.velocity != 0:
                notes.append(msg.note - local_min)  
    return notes
```

```python
def accuracy_class(ground_truth, predicted_labels):
    solo_correct = 0
    total_solo = 0
    window_correct = 0
    total_window = 0
    total_items = 0
    total_correct = 0
    for i in range(len(ground_truth)):
        total_items += 1
        if ground_truth[i] == 0:
            total_window += 1
            if ground_truth[i] == predicted_labels[i]:
                window_correct += 1
                total_correct += 1
        else:
            total_solo += 1
            if ground_truth[i] == predicted_labels[i]:
                solo_correct += 1
                total_correct += 1
    # print("Total correct rate: {}".format((solo_correct+window_correct)/(total_solo+total_window)))
    print("Total correct rate: {}/{}({})".format(total_correct, total_items, total_correct/total_items))
    print("Solos correct classification rate: {}/{}({})".format(solo_correct, total_solo, solo_correct/total_solo))

    print("Window/Non-solo correct classification rate: {}/{}({})".format(window_correct, total_window, window_correct/total_window))
```

```python
def calcPF(seqs):
    pitchcounts = {}
    for seq in seqs:
        for pitch in seqs[seq]:
            if pitch not in pitchcounts.keys():
                pitchcounts[pitch] = 0
            pitchcounts[pitch] += 1
    return pitchcounts

def calcPFdict(seqs):
    PFdict = {}
    totaldiffs = 0.0
    for seq in seqs:
        diffpitches = []
        for pitch in seqs[seq]:
            if pitch not in diffpitches:
                diffpitches.append(pitch)
        if len(seqs[seq]) == 0:
            PFdict[seq] = 0
            continue
        PFval = float(len(diffpitches))/len(seqs[seq])
        PFdict[seq] = PFval
        totaldiffs += PFval
        #print("{} {}/{}".format(totaldiffs, len(diffpitches), len(seqs[seq])))
    return PFdict

def calcPFseq(seqs):
    totaldiffs = 0.0
    for seq in seqs:
        diffpitches = []
        for pitch in seqs[seq]:
            if pitch not in diffpitches:
                diffpitches.append(pitch)
        if len(seqs[seq]) == 0:
            continue
        totaldiffs += float(len(diffpitches))/len(seqs[seq])
        #print("{} {}/{}".format(totaldiffs, len(diffpitches), len(seqs[seq])))
    return totaldiffs/len(seqs)
```

```python
def calcDelta(seqs):
    pitchcounts = {}
    for seq in seqs:
        for pitch in seqs[seq]:
            if pitch not in pitchcounts.keys():
                pitchcounts[pitch] = 0
            pitchcounts[pitch] += 1
    return pitchcounts


def calcDeltaDict(seqs):
    totaldiffs = 0.0
    delta_dict = {}
    for seq in seqs:
        diffpitches = []
        for pitch in seqs[seq]:
            if pitch not in diffpitches:
                diffpitches.append(pitch)
        if len(seqs[seq]) == 0:
            delta_dict[seq] = 0
            continue
        highest = 0
        lowest = 128
        for pitch in diffpitches:
            if highest < pitch:
                highest = pitch
            if lowest > pitch:
                lowest = pitch
        delta = highest-lowest
        delta_dict[seq] = delta
        totaldiffs += delta
        #print("{} {}/{}".format(totaldiffs, len(diffpitches), len(seqs[seq])))
    return delta_dict

def calcDeltaSeq(seqs):
    totaldiffs = 0.0
    for seq in seqs:
        diffpitches = []
        for pitch in seqs[seq]:
            if pitch not in diffpitches:
                diffpitches.append(pitch)
        if len(seqs[seq]) == 0:
            continue
        highest = 0
        lowest = 128
        for pitch in diffpitches:
            if highest < pitch:
                highest = pitch
            if lowest > pitch:
                lowest = pitch
        totaldiffs += highest - lowest
        #print("{} {}/{}".format(totaldiffs, len(diffpitches), len(seqs[seq])))
    return totaldiffs/len(seqs)
```

```python
def calcAverageNoteCount(seqs):
    totalnotes = 0
    for seq in seqs:
        totalnotes += len(seqs[seq])
    return totalnotes/112
```
