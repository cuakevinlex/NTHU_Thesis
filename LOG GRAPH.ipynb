```python
import pickle
import matplotlib.pyplot as plt
```

```python
folder = "dynamic_mag_models/2enc_2nd"
with open(folder + "/train_loss_log.pkl", 'rb') as f:
    x = pickle.load(f)
with open(folder + "/valid_loss_log.pkl", 'rb') as f:
    y = pickle.load(f)
```

```python
len(y)
```

```python
def draw_result(lst_iter, lst_loss, lst_acc, title):
    plt.plot(lst_iter, lst_loss, '-b', label='train loss')
    plt.plot(lst_iter, lst_acc, '-r', label='valid loss')

    plt.xlabel("n iteration")
    plt.legend(loc='upper left')
    plt.title(title)

    # save image
    plt.savefig(title+".png")  # should before show method

    # show
    plt.show()
```

```python
draw_result(range(2000),x,y, "Train and Loss of Accompaniment Generation Model")
```
